```python
import pandas as pd
import numpy as np
import math
import time
import datetime
```

# Import Indicator Data

```python
df_ind = pd.read_csv('combined_indicators.csv')
df_ind.head()
```

```python
df_ind_t = df_ind[['DATE', 'GDP', 'REAL_USER_COST', 'UNEMPLOYMENT_RT']]
```

```python
df_ind_t.columns = ['Date', 'GDP', 'REAL_USER_COST', 'UNEMPLOYMENT_RT']
```

```python
df_ind_t.head()
```

```python
df_ind_t['Date'] = df_ind_t['Date'].apply(lambda x: str(x))
```

```python
df_ind_t.iloc[0,0]
```

# Next data

```python
df_sec = pd.read_csv('raw-data.csv')
```

```python
df_sec_2 = df_sec.drop(columns = ['Unnamed: 0'])
```

```python
df_sec_2
```

```python
df_sec_2.iloc[0,0]
```

# Move returns to quarterly basis

# Combined quarterly return dataframe with quarterly indicator data

```python
joined_raw = df_ind_t.merge(df_sec_2, on=['Date'], how='outer')
```

## Sort by Date ascending

```python
joined_raw_sort = joined_raw.sort_values(by='Date', ascending=True)
```

```python
joined_raw_sort.iloc[:20,:]
```

# Forward fill all sector return columns

```python
#copy
joined_ffil = joined_raw_sort
```

```python
joined_ffil.columns
```

```python
joined_ffil['XLB'] = joined_ffil['XLB'].ffill()
joined_ffil['XLE'] = joined_ffil['XLE'].ffill()
joined_ffil['XLF'] = joined_ffil['XLF'].ffill()
joined_ffil['XLI'] = joined_ffil['XLI'].ffill()
joined_ffil['XLK'] = joined_ffil['XLK'].ffill()
joined_ffil['XLU'] = joined_ffil['XLU'].ffill()
joined_ffil['XLV'] = joined_ffil['XLV'].ffill()
joined_ffil['XLY'] = joined_ffil['XLY'].ffill()
joined_ffil['GSPC'] = joined_ffil['GSPC'].ffill()
```

# Trim up to only yield quarterly date columns

```python
dates = df_ind_t.Date
```

```python
final_output = joined_ffil[joined_ffil.Date.isin(dates)].iloc[4:,:]
```

```python
final_output.to_csv('quarterly_final.csv')
```
